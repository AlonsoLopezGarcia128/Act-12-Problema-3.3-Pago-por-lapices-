```python
# En caso de ser verdadero (Si)
# Definir pago por cantidad de lapices
# Valores de entrada x = Cantidad de lapices
# Decidir si X es igual o mayor a 1000
# Si es mayor o igual definir PAG = X * 0.85
# Si es menor definir PAG = X * 0.90
# PAG = Pago total
X = int(input("Ingrese la cantidad de lapices: "))
if X >= 1000:
    PAG = X * 0.85
    print("El pago total es: ", PAG)
else:
    PAG = X * 0.90
    print("El pago total es: ", PAG)
print("Programa realizado por Alonso Lopez Garcia NL = 31")
```
